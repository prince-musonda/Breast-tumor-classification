```python
import torch
from torch import nn
from torchvision import transforms
from PIL import Image

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

# Load Resnet101 model, and remake the classification  layer

```python
from torchvision import models
torch.manual_seed(42)


lungCancer_model = models.resnet101(weights=models.ResNet101_Weights.DEFAULT)
lungCancer_model.classifier = nn.Sequential(nn.Linear(in_features=2048, out_features=4))
```

```python

from google.colab import drive
drive.mount('/content/drive')
```

# load in the saved/trained weights

```python
lungCancer_model_weights = torch.load('/content/drive/MyDrive/lung.pth')
lungCancer_model.load_state_dict(lungCancer_model_weights)
```

```python
transform_image = models.ResNet101_Weights.DEFAULT.transforms()

lungCancer_model.to(device)
```

# Gradio set up

```python
def prediction(img):
  # remove alpha channel if any
  img.convert('RGB')
  img = transform_image(img).to('cuda').unsqueeze(0)
  lungCancer_model.eval()
  with torch.inference_mode():
    pred_prob = torch.softmax(lungCancer_model(img), dim=1).squeeze()
    print(pred_prob)

  predication_label_n_prob = {'Adenocarcinoma Detected' : pred_prob[0],
                              'large cell carcinoma Detected': pred_prob[1],
                              'No Lung Cancer detected': pred_prob[2],
                              'squamous cell carcinoma Detetcted': pred_prob[3]
                              }
  return predication_label_n_prob
```

```python
import gradio as gr
demo = gr.Interface(fn =prediction,
                    inputs=gr.Image(type='pil'),
                    outputs = gr.Label(num_top_classes=1,label='predictions'),
                    title='iCare.AI Lung Cancer Detection and Diagnosis system',
                    description='Please provide a chest CT scan')
```

```python
demo.launch(debug=True, share=True)
```
