```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras.layers import Normalization
from tensorflow.keras import Sequential
```

```python
import pandas as pd
```

```python
breast_cancer_data = pd.read_csv('Breast_cancer_data.csv')
breast_cancer_data.head()
```

```python
X = breast_cancer_data[['mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness',]].values
```

```python
y = breast_cancer_data[['diagnosis']].values
```

**Split the data**

```python
x_train = X[0:500]
y_train = y[0: 500]
x_test = X[500:]
y_test= y[500:]
```

# Feature Scaling: **normalize our data**

```python
normalizer = Normalization(axis=1)
normalizer.adapt(x_train)
normalized_x_train = normalizer(x_train)
```

# **Build model**

```python
bCancerNet = Sequential([
    Dense(units=6,activation='relu'),
    Dense(units=5, activation='relu'),
    # Dense(units=4, activation='relu'),
    #  Dense(units=3, activation='relu'),
     Dense(units=2, activation='relu'),
    Dense(units=1, activation='sigmoid')
])
```

```python
bCancerNet.compile(optimizer='Adam',loss='BinaryCrossentropy')
```

```python
bCancerNet.fit(normalized_x_train,y_train, epochs=195)
```

# Evaluate our model's performance

```python
normalized_x_test = normalizer(x_test)
bCancerNet.evaluate(normalized_x_test, y_test)
```

**Make prediction with patient number 1 who has cancer**

```python
patient = normalized_x_test[0]
print(y_test[0])
bCancerNet.predict(tf.reshape(patient, shape=(1,5)))
```
